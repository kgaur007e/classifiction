```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
#loading data set
iris=load_iris()
iris
```

```python
labels=iris.target #label name       
```

```python
iris.feature_names #features name
```

```python
features=iris.data #actual data of feature
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

#only setosa data
setosa=features[:50]
plt.xlabel('sl')
plt.ylabel('sw')
plt.title('setosa data')

plt.scatter(setosa[:,0],setosa[:,1],c='red')
plt.scatter(setosa[:,2],setosa[:,-1],c='green',marker="*")
plt.show()
```

```python
#lets split Data into traning and testing
train_X,test_X,train_Y,test_Y=train_test_split(features,labels,test_size=0.2)
```

```python

#calling model
model=DecisionTreeClassifier()#calling classifier
#traning model
desc_trained=model.fit(train_X,train_Y)
```

```python
#predicting
predicted_output=desc_trained.predict(test_X)
```

```python
#actual output
actual_output=test_Y
actual_output
```

```python
#accuracy_score
from sklearn.metrics import accuracy_score
```

```python
#finding
score=accuracy_score(test_Y,predicted_output)
```

```python
print("accuracy of decision tree",score)
```

```python
#lets plot tree stucture
plt.figure(figsize=(12,8))#defining figure size
plot_tree(desc_trained,feature_names=iris.feature_names,class_names=iris.target_names.tolist(),filled=True)
plt.title("DecisionTree")
plt.show()
```
